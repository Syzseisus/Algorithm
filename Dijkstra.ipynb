````python
# Colab에서 drive mount 및 input.txt 파일 경로 설정
from google.colab import drive
drive.mount('/content/drive')
input_path = "/content/drive/MyDrive/Colab Notebooks/Junior_alogrithm_Graph/input.txt"

def dijkstra(n, weights):
    ### 초기화
    # start == 맨 처음 시작 node
    start = 1

    # node_dist == {node:start ~ node 거리}
    node_dist = {i: float('inf') for i in range(2, len(weights[0]))}
    node_dist[start] = 0
    num_node = len(node_dist) # node 개수

    # search_node == 탐색해야할 node들
    search_node = list(range(1,num_node+1))

    # parent == 선택될 next_node의 parent node
    # 초기값 None, 어차피 첫 번째 반복문 때 갱신됨
    parent = {i:None for i in range(1,num_node+1)}

    # answer == 총 거리 : 0에서 시작
    answer = 0


    ### 현재 node에 따른 변수
    # curr == 현재 node, 처음은 start (int)
    curr = 1
    search_node.remove(curr) # search_node에서 curr 하나 씩 삭제

    # from_start_dist == 시작 node에서 현 node까지 거리, 처음은 0
    from_start_dist = 0


    ### 반복문: search_node 다 빌 때까지
    while search_node:
        # connection == curr와 연결된 node들의 거리 (node:dist, dict)
        connection = {i:weights[curr][i] for i in range(1,num_node+1) if (weights[curr][i] != float('inf')) & (i != curr)}
        print("connection:", connection)

        # node_dist 업데이트: 현재 값 / 시작 ~ curr ~ node를 비교해서 작은 값으로 업데이트
        for node in connection:
            new_dist = connection[node] + from_start_dist
            # node 갱신 시에 parent도 변경
            if node_dist[node] > new_dist:
                node_dist[node] = new_dist
                parent[node] = curr

        # min_val   == node_dist에서 가장 작은 value
        # next_node == node_dist에서 min_val을 value로 갖는 key 중
        #              serach_node에 없는 값으로 다음 node 지정
        min_val = min({key:node_dist[key] for key in node_dist if key in search_node}.values())
        for node in node_dist:
            if node in search_node:
                if node_dist[node] == min_val:
                    next_node = node

        # answer 업데이트: next_node의 parent에서 next_node까지의 거리
        answer += weights[parent[next_node]][next_node]

        # node 변경
        curr = next_node
        from_start_dist = node_dist[curr]
        search_node.remove(next_node) # search_node에서 curr 하나 씩 삭제


    return answer

if __name__ == "__main__":
    with open(input_path, "r") as f:
        num_cases = int(f.readline())
        for i in range(num_cases):
            n = int(f.readline())
            weights = [[0 if j == k else float('inf') for j in range(n+1)] for k in range(n+1)]

            for j in range(n):
                node = [int(x) for x in f.readline().split()]
                k, m = node[0], node[1]
                for l in range(m):
                    weights[k][node[l*2+2]] = node[l*2+3]

            print(dijkstra(n, weights))
````
